## Querying and sorting rows

````python
## query all rows having new_cases > 1000

covid_df_copy[covid_df_copy['new_cases'] > 1000]
````

````python
from IPython.display import display

# display all rows
with pd.option_context('display.max_rows', 100):
    display(covid_df[covid_df['new_cases'] > 1000])
````

````python
total_cases = covid_df['new_cases'].sum()
initial_tests = 935310
total_tests = initial_tests + covid_df.new_tests.sum()

positive_rates = total_cases / total_tests

positive_rates
````

````python
result_df = covid_df[covid_df['new_cases'] / covid_df['new_tests'] > positive_rates]

result_df
````

````python
## Add column positive_rates into covid_df_copy
covid_df_copy['positive_rates'] = covid_df['new_cases'] / covid_df['new_tests']
````

````python
covid_df_copy
````

````python
## Drop column positive_rates
covid_df_copy.drop(columns = 'positive_rates', inplace = True)

covid_df_copy
````

````python
## Find 5 rows with most deaths
covid_df.sort_values('new_deaths', ascending = False).head(5)
````

````python
# Change values at a specific data rows
covid_df_copy.at[169, 'new_cases'] = 300.6

covid_df_copy.loc[169]
````

````python
## Convert object to date type
covid_df_copy['date'] = pd.to_datetime(covid_df_copy['date'])

covid_df_copy['date']
````

````python
# Add year,month,day,weekday column to covid copy dataframe
covid_df_copy['year'] = pd.DatetimeIndex(covid_df_copy['date']).year
covid_df_copy['month'] = pd.DatetimeIndex(covid_df_copy['date']).month
covid_df_copy['day'] = pd.DatetimeIndex(covid_df_copy['date']).day
covid_df_copy['weekday'] = pd.DatetimeIndex(covid_df_copy['date']).weekday

covid_df_copy
````

````python
# Calculate sum of all new_cases, new_deaths and new_tests in May
total = covid_df_copy[covid_df_copy['month'] == 5][['new_cases', 'new_deaths', 'new_tests']].sum()

total
````

## Grouping and aggregation

````python
# Group by month and calculate sum of it

new_group_df = covid_df_copy.groupby('month')[['new_cases', 'new_deaths', 'new_tests']].sum()

new_group_df
````

````python
# Calculate cumulative sum of last 5 rows by new_cases
totalSumTest = covid_df_copy['new_cases'].tail(5).cumsum()

totalSumTest
````

````python
covid_df_copy['total_cases'] = covid_df_copy['new_cases'].cumsum()

covid_df_copy
````

## Merging data from multiple sources

````python
import urllib.request

urllib.request.urlretrieve('https://gist.githubusercontent.com/aakashns/8684589ef4f266116cdce023377fc9c8/raw/99ce3826b2a9d1e6d0bde7e9e559fc8b6e9ac88b/locations.csv', 
            'locations.csv')

location_df = pd.read_csv('locations.csv')

location_df
````

````python
## Add column location into covid_df_copy to merge
covid_df_copy['location'] = 'Italy'

covid_df_copy
````

````python
merge_df = covid_df_copy.merge(location_df, on = 'location')

merge_df
````

````python
# Calculate cases per million
merge_df['cases_per_million'] = merge_df['total_cases'] * 1e6 / merge_df['population']

merge_df
````

## Writing data back to files

````python
result_df = merge_df[['date', 'new_cases', 'new_deaths', 'new_tests', 'total_cases', 'cases_per_million']]

result_df
````

````python
# Write result_df to files
result_df.to_csv('results.csv', index = None)
````
